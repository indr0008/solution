```python
#we have to change it into a single row containing all information,
#then concat all the data from the multiple json into a single dataframe
```

```python
#create a function to change the json into a single row dataframe
def takedata(filename):
    dataframe = pd.DataFrame(json.load(open(filename)))
    dataframe2 = dataframe.T
    dataframe2 = dataframe2.tail(1)
    dataframe2.reset_index(drop=True, inplace=True)
    dataframe = dataframe.head(1)
    dataframe.reset_index(drop=True, inplace=True)
    dataframe = dataframe[['id','op','ts']]
    dataframe3 = pd.concat([dataframe, dataframe2], axis=1)
    return dataframe3
```

```python
file_list = []   # define file_list to save all json files
for subdir, dirs, files in os.walk(working_directory):
    for file in files:
        if file.endswith('.json'):
            file_list.append(file)
```

```python
#for loop
savings = pd.DataFrame()
for input_file in file_list:
    test1 = takedata(input_file)
    savings = pd.concat([savings,test1])
```

```python
savings #check result
```

```python
#fill the NA with preceding values, because here if things are not updated, it should stay the same 
savings_filled = savings.fillna(method='ffill')
```

```python
#to merge the tables, we merge the timestamp while making sure that the timelines are correct
merge2 = pd.merge(cards_filled,savings_filled,left_on=['ts','op','status'],right_on=['ts','op','status'],how='outer',indicator=True)
merge2 = merge2.sort_values(by=['ts'])
merge2 = merge2.fillna(method='ffill')
merge2.reset_index(drop=True, inplace=True)
```
